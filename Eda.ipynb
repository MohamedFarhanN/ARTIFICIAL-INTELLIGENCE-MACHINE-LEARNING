```python
# Chart - 1 Histogram visualization code for distribution of numerical variables
# Create a figure with subplots
plt.figure(figsize=(8, 6))
plt.suptitle('Distribution of Iris Flower Measurements', fontsize=14)
# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1) # Subplot 1 (Top-Left)
plt.hist(df['sepal.length'])
plt.title('Sepal Length Distribution')
plt.subplot(2, 2, 2) # Subplot 2 (Top-Right)
plt.hist(df['sepal.width'])
plt.title('Sepal Width Distribution')
plt.subplot(2, 2, 3) # Subplot 3 (Bottom-Left)
plt.hist(df['petal.length'])
plt.title('Petal Length Distribution')
plt.subplot(2, 2, 4) # Subplot 4 (Bottom-Right)
plt.hist(df['petalwidth']) 
plt.title('Petal Width Distribution')
# Display the subplots
plt.tight_layout() # Helps in adjusting the layout
plt.show()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="petal.length", data=df)
```

```python
ax = sns.boxplot(x="Species", y="petal.length", data=df)
ax = sns.stripplot(x="Species", y="petal.length", data=df, jitter=True, edgecolor="gray")
```

```python
# Create the figure
plt.figure(figsize=(8, 6))
species = df['Species'].unique()
colors = ['red', 'green', 'blue']
# Scatter plot for each species
for i in range(len(species)):
 x = df[df['Species'] == species[i]]
 plt.scatter(x['sepal.length'], x['sepal.width'], c=colors[i], label=species[i])
# Add labels and legend
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('sepal length vs width by Species')
plt.legend()
# Show plot
plt.show()
```

```python
# Calculate correlation
corr = iris.drop('species', axis=1).corr()
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
# Plot distributions by species
sns.boxplot(x='species', y='sepal_length', data=iris, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Sepal Length by Species')
sns.boxplot(x='species', y='sepal_width', data=iris, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Sepal Width by Species')
sns.boxplot(x='species', y='petal_length', data=iris, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Petal Length by Species')
sns.boxplot(x='species', y='petal_width', data=iris, ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Petal Width by Species')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='petal_length', data=iris, inner='quartile', palette='pastel')
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(iris, 'species', colormap='viridis')
plt.title('Andrews Curves for Iris Dataset')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(iris, 'species', colormap='viridis')
plt.title('Parallel Coordinates Plot for Iris Dataset')
plt.show()
```
