```python
import pandas as pd
#from dictionery
data={'Name':['alice','bob'],'age':[25,30]}
df=pd.DataFrame(data)
print(df)

```

```python
# From list of lists 
lst = [['Alice', 25], ['Bob', 30]] 
df = pd.DataFrame(lst, columns=['Name', 'Age']) 
print(df)
```

```python
#Merging And Joining
df1 = pd.DataFrame({
    'key': [1, 2, 3],
    'Name': ['Hari', 'Arya', 'Vishnu']
})
df2 = pd.DataFrame({
    'key': [1, 2, 4],
    'Age': [25, 30, 35]
})
pd.merge(df1, df2, on='key')
```
